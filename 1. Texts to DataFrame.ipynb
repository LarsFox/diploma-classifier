```python
from os import listdir
```

```python
def files(path):
    return [f for f in listdir(path)]
```

```python
meduza = files('../diploma-news-crawler/.tmp/meduza/news')
```

```python
rt = files('../diploma-news-crawler/.tmp/rt/all-news')
```

```python
tass = files('../diploma-news-crawler/.tmp/tass')
tass = {
    'obschestvo': files('../diploma-news-crawler/.tmp/tass/obschestvo'),
    'ekonomika': files('../diploma-news-crawler/.tmp/tass/ekonomika'),
    'politika': files('../diploma-news-crawler/.tmp/tass/politika'),
    'sport': files('../diploma-news-crawler/.tmp/tass/sport'),
    'kultura': files('../diploma-news-crawler/.tmp/tass/kultura'),
}
tass
```

```python
texts = []

for category in tass:
    for filename in tass[category]:
        with open(f'../diploma-news-crawler/.tmp/tass/{category}/{filename}', 'r') as target:
            src = target.read()

        url = f'tass.ru/{category}/{filename[:-4]}'
        texts.append(
            {'filename': filename, 'category': category, 'url': url, 'text': src}
        )

```

## Creating DF

```python
import pandas as pd

from helpers import dump_pickle


dfTass = pd.DataFrame(texts)
dfTass
```
