## Removing non-letters

```python
df['trimmed'] = df['text'].apply(trim)
df.sample()['trimmed'].iloc[0]
```

## Lemmatization

```python
df['lemmas'] = df['trimmed'].apply(lemmatize)
df.sample()['lemmas'].iloc[0]
```

## Removing stop words

```python
df['stopped'] = df['lemmas'].apply(stopwordize)
df.sample()['stopped'].iloc[0]
```

## Finalization

```python
df['final'] = df['stopped'].apply(trim)
df.sample()['final'].iloc[0]
```
