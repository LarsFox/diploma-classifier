```python
rfc = load_pickle('tass/classifiers/rfc')
svm = load_pickle('tass/classifiers/svm')
knn = load_pickle('tass/classifiers/knn')
mnbc = load_pickle('tass/classifiers/mnbc')
lr = load_pickle('tass/classifiers/lr')
gbm = load_pickle('tass/classifiers/gbm')
```

```python
tfidf = load_pickle('tass/tfidf')
```

```python
features_test = load_pickle('tass/features_test')
labels_test = load_pickle('tass/labels_test')
```

```python
CATEGORIES = {
    0: 'obschestvo',
    1: 'ekonomika',
    2: 'politika',
    3: 'sport',
    4: 'kultura'
}
```

```python
def predict_all(text):
    features = tfidf.transform(pd.Series(finalize(text))).toarray()
    print('RF:', CATEGORIES.get(rfc.predict(features)[0]), rfc.predict_proba(features))
    print('SVM:', CATEGORIES.get(svm.predict(features)[0]), svm.predict_proba(features))
    print('kNN:', CATEGORIES.get(knn.predict(features)[0]), knn.predict_proba(features))
    print('MNBC:', CATEGORIES.get(mnbc.predict(features)[0]), mnbc.predict_proba(features))
    print('LR:', CATEGORIES.get(lr.predict(features)[0]), lr.predict_proba(features))
    print('GBM:', CATEGORIES.get(gbm.predict(features)[0]), gbm.predict_proba(features))
```
